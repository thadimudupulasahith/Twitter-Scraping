```python
scraper = sntwitter.TwitterSearchScraper("#python")
tweets = []
n_tweets = 1_000
for i, tweet in tqdm(enumerate(scraper.get_items())):
    data = [tweet.date,tweet.id,tweet.rawContent,tweet.user.username,]
    tweets.append(data)
    if i > n_tweets:
        break
    
```

```python
len(tweets)
```

```python
tweet_df = pd.DataFrame(tweets,columns=["date","id","content","username"])
```
