```python
import pandas as pd
import snscrape.modules.twitter as sntwitter

# Set search parameters
search_words = 'keyword OR hashtag'
start_date = '2021-01-01'
end_date = '2021-01-31'
max_tweets = 1000

# Define dataframe columns
columns = ['date', 'id', 'url', 'content', 'username', 'reply_count', 'retweet_count', 'language', 'source', 'like_count']

# Create an empty dataframe to store scraped data
df = pd.DataFrame(columns=columns)

# Use snscrape to scrape tweets
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{search_words} since:{start_date} until:{end_date}').get_items()):
    # Break loop if max number of tweets has been reached
    if i >= max_tweets:
        break
    # Extract tweet data and add to dataframe
    df.loc[i] = [tweet.date, tweet.id, tweet.url, tweet.content, tweet.user.username, tweet.replyCount, tweet.retweetCount, tweet.lang, tweet.sourceLabel, tweet.likeCount]
    
# Print the dataframe
print(df.head())
```

```python
import streamlit as st
import pandas as pd
from datetime import datetime, timedelta
from snscrape.modules import twitter
import pymongo

# Set up MongoDB client and database
client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['twitter_data']
collection = db['tweets']

# Define function to scrape Twitter data
def scrape_twitter_data(keyword, start_date, end_date, tweet_count):
    tweets = []
    for tweet in twitter.TwitterSearchScraper(f'{keyword} since:{start_date} until:{end_date}').get_items():
        if len(tweets) >= tweet_count:
            break
        tweets.append({
            'Date': tweet.date.strftime('%Y-%m-%d %H:%M:%S'),
            'ID': tweet.id,
            'URL': tweet.url,
            'Content': tweet.content,
            'User': tweet.user.username,
            'Reply Count': tweet.replyCount,
            'Retweet Count': tweet.retweetCount,
            'Language': tweet.lang,
            'Source': tweet.source,
            'Like Count': tweet.likeCount
        })
    return pd.DataFrame(tweets)

# Define function to upload data to MongoDB
def upload_to_mongodb(data, keyword):
    collection.insert_one({
        'Scraped Word': keyword,
        'Scraped Date': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        'Scraped Data': data.to_dict('records')
    })

# Set up Streamlit app
st.title('Twitter Data Scraper')

# Get user input for keyword, start date, end date, and tweet count
keyword = st.text_input('Enter a keyword or hashtag:')
start_date = st.date_input('Select a start date:')
end_date = st.date_input('Select an end date:')
tweet_count = st.number_input('Enter the maximum number of tweets to scrape:', min_value=1, max_value=1000, value=100)

# Convert date inputs to string format
start_date_str = start_date.strftime('%Y-%m-%d')
end_date_str = end_date.strftime('%Y-%m-%d')

# Scrape Twitter data and display in table
if st.button('Scrape Data'):
    st.write(f'Scraping Twitter data for "{keyword}" from {start_date_str} to {end_date_str}...')
    data = scrape_twitter_data(keyword, start_date_str, end_date_str, tweet_count)
    st.write(f'{len(data)} tweets scraped:')
    st.dataframe(data)

# Add button to upload data to MongoDB
if st.button('Upload to MongoDB'):
    upload_to_mongodb(data, keyword)
    st.write('Data uploaded to MongoDB.')

# Add buttons to download data in CSV and JSON formats
if st.button('Download as CSV'):
    csv = data.to_csv(index=False)
    st.download_button('Download CSV', data=csv, file_name='twitter_data.csv', mime='text/csv')

if st.button('Download as JSON'):
    json = data.to_json(orient='records')
    st.download_button('Download JSON', data=json, file_name='twitter_data.json', mime='application/json')
```
